```python
from  sentinelsat  import  SentinelAPI ,  read_geojson ,  geojson_to_wkt 
```

```python
from datetime import datetime
```

```python
#Conexión API 
api = SentinelAPI('monitoreogeografia', 'monitoreogeo08', 'https://scihub.copernicus.eu/dhus')
```

```python
# Uso de un archivo geojson para el área de consulta
footprint = geojson_to_wkt(read_geojson('C:/JSONEDOMEX/EdoMex.json'))
```

```python
products = api.query(footprint,
                     date=('20230701', '20230710'),
                     platformname = 'Sentinel-2',
                     cloudcoverpercentage = (0, 30))
            
```
